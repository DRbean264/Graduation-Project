```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_source = 'ECNU'
```

```python
def Plot_Learning_Curve(history,metric_name,include_val=False,save=False,savename=None):
    plt.plot(history[metric_name])
    plt.plot(history['val_{}'.format(metric_name)])
    plt.legend(['Training', 'Testing'], loc='upper right')
        
    plt.title('{} vs. epochs'.format(metric_name.title()), fontsize=18)
    plt.ylabel(metric_name.title(), fontsize=16)
    plt.xlabel('Epoch', fontsize=16)
    plt.ylim(0,0.040)
    plt.tick_params(labelsize=14)
    
    if save:
        plt.tight_layout()
        plt.savefig("./Plot-Analysis/Learning Curve/Learning_Curve_{}/{}".format(dataset_source,savename),dpi=300)
    
    plt.show()
```

```python
with open("./Plot-Analysis/Learning Curve/Learning_Curve_ECNU/Histroy-ECNU-CNN-DNN.txt") as f:
    history = eval(f.read())
    
Plot_Learning_Curve(history,'loss',include_val=True,save=False,savename='CNN-DNN')
```

```python
with open("./Plot-Analysis/Learning Curve/Learning_Curve_ECNU/Histroy-ECNU-CNN-CNN.txt") as f:
    history = eval(f.read())
    
Plot_Learning_Curve(history,'loss',include_val=True,save=False,savename='CNN-CNN')
```
